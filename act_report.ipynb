## WeRateDogs Wrangling Project - Analysing and Visualing Data

For the purpose of this project, 3 insights were generated and 1 visualization was done after the wrangling process.
Insights are very important and the use of visualizations can also be used in generating these insights. 

Below are the details of the analysis performed for this project:

### Analysis of Dog Stages

Here, we are concerned about the different dog stages and the most occuring and least occuring dog stage in the data set. A bar chart was used to visualize the data and it was observed that the pupper stage is the most common stage among all the dogs present in the data set. It can also be noticed that the doggoflopper is the least occuring which indicates that not so many dogs are in this stage. 

### Correlation between Retweet_Count and Favorite_Count

Another insight generated is that there is a positive relationship between retweet_count and favorite_count in this data set. This implies that an increase in the retweet_count will also leade to an increase in the favorite_count and vice-versa. This insight was revealed through the use of a scatter plot which clearly shows this correlation. However, there are other ways that this insight can be generated.
